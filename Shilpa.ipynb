```python
# Dependencies and SetUp 

import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import requests
import os 
import gmaps 
```

```python
df = pd.read_csv('nobel prize.csv')
df1=df.dropna()
```

```python
# Which Gender has the most Nobel Prize Winner 

gender = df1['Sex'].value_counts()
x_list=[]
y_list=[]
j=0
for i in gender.index:
    x_list.append(i)
    selectTemp=df[df['Sex']==i]
    y_list.append(selectTemp["Sex"].count())
    j+=1
    if j==100:
        break
plt.figure(figsize=(8,5))
plt.bar(x_list,y_list,color='#0504aa',alpha=.5,align="center",width= .25)
plt.xlabel('Gender')
plt.ylabel("Number of Winners")
plt.xticks(rotation = 45)
plt.tight_layout()
output_img = 'shilpa_output/'+'Most Gender Wins BarPlot' +'.png'
plt.savefig(output_img, dpi=300,bbox_inches='tight')

```

```python
# Same as above but plotted in pie chart instead 

pies = gender.index
pie_votes = gender
explode = (0.3, 0) 
plt.figure(figsize=(5,5))
plt.pie(pie_votes, labels=pies,explode=explode, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
output_img = 'shilpa_output/'+'Most Gender Wins PieChart' +'.png'
plt.savefig(output_img, dpi=300,bbox_inches='tight')
gender
```
