```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Create Dataframe
df= pd.read_csv('nobel prize.csv')

# Remove Organizations
df = df[df["Laureate Type"] != "Organization"]

# Calculate new colume containing  United States and Other Countries
df.loc[(df['Birth Country'] == "United States of America")  | (df['Organization Country'] == "United States of America"), 'United States'] = 'United States'
df.loc[df['United States'] != 'United States', 'United States'] = 'Other Country' 

# Calculate STEM and non-STEM categories
df.loc[(df['Category'] == "Literature")  | (df['Category'] == "Peace")  | (df['Category'] == "Economics"), 'STEM'] = 'STEM'
df.loc[df['STEM'] != 'STEM', 'STEM'] = 'Non-STEM' 

#df.to_csv('file_name.csv', index=False)

# Create female Dataframes
female_df = df[(df["Sex"]=="Female") & ((df["Category"]=="Physics") | (df["Category"]=="Medicine"))]
female_all_df = df[(df["Sex"]=="Female") ]

# Create male Dataframes
male_df = df[(df["Sex"]=="Male") & ((df["Category"]=="Physics") | (df["Category"]=="Medicine"))] 
male_all_df = df[(df["Sex"]=="Male") ]

# Create male/female Dataframes with Physics and Medicine only
male_female_df = df[((df["Category"]=="Physics") | (df["Category"]=="Medicine"))]
```

```python
# Aggregate Data
female_series = female_df["Category"].value_counts().rename_axis('Category').reset_index(name='Count')

# Set variables
x_axis = female_series["Category"]
y_axis = female_series["Count"]

# Create Graph
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",shadow=True,startangle=45)
plt.title("Female Nobel Prize Winners \nMedicine vs. Physics")

# Display Graph
plt.show()
```

```python
# Aggregate Data
female_all = female_all_df["Category"].value_counts().rename_axis('Category').reset_index(name='Count')

# Set variables
x_axis = female_all["Category"]
y_axis = female_all["Count"]

# Create Graph
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",shadow=True,startangle=45)
plt.title("Female Nobel Prize Winners \nAll Categories")

# Display Graph
plt.show()
```

```python
# Aggregate Data
female_all_df = df[(df["Sex"]=="Female") ]
female_country = female_all_df["United States"].value_counts().rename_axis('United States').reset_index(name='Count')

# Set variables
x_axis = female_country["United States"]
y_axis = female_country["Count"]

# Create Graph
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",shadow=True,startangle=45)
plt.title("Female Nobel Prize Winners \nUnited States vs. Other Countries")

# Display Graph
plt.show()
```

```python
# Aggregate Data
male_all_df = df[(df["Sex"]=="Male") ]
male_country = male_all_df["United States"].value_counts().rename_axis('United States').reset_index(name='Count')

# Set variables
x_axis = male_country["United States"]
y_axis = male_country["Count"]

# Create Graph
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",shadow=True,startangle=45)
plt.title("Male Nobel Prize Winners \nUnited States vs. Other Countries")

# Display Graph
plt.show()
```

```python
# Aggregate Data
female_all_df = df[(df["Sex"]=="Female") ]
female_STEM = female_all_df["STEM"].value_counts().rename_axis('STEM').reset_index(name='Count')

# Set variables
x_axis = female_STEM["STEM"]
y_axis = female_STEM["Count"]

# Create Graph
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",shadow=True,startangle=45)
plt.title("Female Nobel Prize Winners \nSTEM vs Non-STEM")

# Display Graph
plt.show()
```
